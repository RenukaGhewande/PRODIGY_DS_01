## Task 01 – Data Visualization

**Objective:**  
To visualize the distribution of a variable using a bar graph.

**Description:**  
In this task, a real-world CSV dataset was loaded using Pandas.  
A bar graph was created to visualize the population distribution of the top 10 countries for the year 2020 using Python and Matplotlib.

**Tools Used:**  
- Python  
- Pandas  
- Matplotlib  
- Jupyter Notebook


```python
import pandas as pd 
df = pd.read_csv(
    r"R:\PRODIGY_DS_01\API_SP.POP.TOTL_DS2_en_csv_v2_34.csv" ,
    skiprows=4
)

df.head()
```

```python
import matplotlib.pyplot as plt

top_10 = df[['Country Name', '2020']].dropna() \
         .sort_values(by='2020', ascending=False) \
         .head(10)

plt.figure(figsize=(10,5))
plt.bar(top_10['Country Name'], top_10['2020'])
plt.title('Top 10 Countries by Population (2020)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
